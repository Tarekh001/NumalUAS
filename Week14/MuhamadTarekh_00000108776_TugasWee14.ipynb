```python
import numpy as np
import matplotlib.pyplot as plt

sampling_rate = 100 
duration = 2  
t = np.linspace(0, duration, int(sampling_rate * duration))  # Array waktu


f1 = 5  
A1 = 3  
phase1 = 3  
signal1 = A1 * np.sin(2 * np.pi * f1 * t + phase1)


f2 = 2  
A2 = 2 
phase2 = -2 
signal2 = A2 * np.sin(2 * np.pi * f2 * t + phase2)


combined_signal = signal1 + signal2

plt.figure(figsize=(8, 6))
plt.plot(t, combined_signal, 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Combined Signal (Sampling Rate = 100 Hz)')
plt.grid(True)
plt.show()
```

```python
sampling_rates = [5, 10, 20, 50, 100]

plt.figure(figsize=(10, 8))

for i, sr in enumerate(sampling_rates, 1):
    t = np.linspace(0, duration, int(sr * duration))
    signal1 = A1 * np.sin(2 * np.pi * f1 * t + phase1)
    signal2 = A2 * np.sin(2 * np.pi * f2 * t + phase2)
    combined_signal = signal1 + signal2
    
    plt.subplot(len(sampling_rates), 1, i)
    plt.plot(t, combined_signal, 'r')
    plt.ylabel('Amplitude')
    plt.title(f'Sampling rate = {sr} Hz')
    plt.ylim(-6, 6) 
    if i == len(sampling_rates):
        plt.xlabel('Time (s)')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
