<h1>Forward Diffrence<h1>

```python
import numpy as np
import matplotlib.pyplot as plt

# Step size
h = 0.1
# Define grid
x = np.arange(0, 2*np.pi, h)
# Compute function
y = np.cos(x)

# Compute vector of forward differences
forward_diff = np.diff(y) / h
# Compute exact solution
x_diff = x[:-1]  # Adjust x to match the length of forward_diff
exact_solution = -np.sin(x_diff)

# Plot solution
plt.figure(figsize=(12, 8))
plt.plot(x_diff, forward_diff, '--', label='Finite differences approximation')
plt.plot(x_diff, exact_solution, label='Exact solution')
plt.legend()
plt.title('Forward Difference Approximation of cos(x) Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Save the plot instead of showing it
plt.savefig('forward_difference_cos.png')
```

<h1>Backward Difference<h1>

```python
import numpy as np
import matplotlib.pyplot as plt

# Step size
h = 0.1
# Define grid
x = np.arange(0, 2*np.pi, h)
# Compute function
y = np.cos(x)

# Compute vector of backward differences
backward_diff = (y[1:] - y[:-1]) / h
# Compute exact solution
x_diff = x[1:]  # Adjust x to match the length of backward_diff
exact_solution = -np.sin(x_diff)

# Compute absolute error and maximum error
error = np.abs(backward_diff - exact_solution)
max_error = np.max(error)
print(f"Maximum error: {max_error}")

# Plot solution
plt.figure(figsize=(12, 8))
plt.plot(x_diff, backward_diff, '--', label='Backward Difference Approximation')
plt.plot(x_diff, exact_solution, label='Exact Solution')
plt.legend()
plt.title('Backward Difference Approximation of cos(x) Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Save the plot
plt.savefig('backward_difference_cos.png')

# Plot error
plt.figure(figsize=(12, 8))
plt.plot(x_diff, error, label='Absolute Error')
plt.legend()
plt.title('Absolute Error of Backward Difference Approximation')
plt.xlabel('x')
plt.ylabel('Error')
plt.grid(True)

# Save the error plot
plt.savefig('backward_difference_error.png')
```
