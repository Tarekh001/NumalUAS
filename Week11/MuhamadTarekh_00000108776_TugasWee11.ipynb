```python
import numpy as np 
import matplotlib.pyplot as plt 
f = lambda x: np.cos(x) 
df_exact = lambda x: -np.sin(x) 

x = np.arange(0, 2*np.pi, 0.01) 
h = x[1] - x[0] 
 
y = f(x) 

df_approx = np.zeros_like(y) 
df_approx[1:] = (y[1:] - y[:-1]) / h 
df_approx[0] = (y[0] - f(x[0] - h)) / h  
 
df_true = df_exact(x) 

error = np.abs(df_true - df_approx) 
max_error = np.max(error) 
print("Maximum error using backward difference:", max_error) 
 
plt.figure(figsize=(10, 6)) 
plt.plot(x, df_approx, 'b--', label='Backward Finite difference approximation') 
plt.plot(x, df_true, 'orange', label='Exact solution') 
plt.xlabel('x') 
plt.ylabel("f'(x)") 
plt.title("Backward Difference Approximation vs Exact Derivative") 
plt.legend() 
plt.grid(True) 
plt.show()
```
