```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.array([
    -3.141592653589793, -2.810898690040254, -2.4802047265182576, -2.14951076298249,
    -1.8188167994467224, -1.4881228359109546, -1.157428872375187, -0.8267349088394194,
    -0.4960409453036516, -0.16534698176788387, 0.16534698176788387, 0.4960409453036516,
    0.8267349088394189, 1.1574288723751867, 1.4881228359109544, 1.8188167994467221,
    2.14951076298249, 2.4802047265182576, 2.810898690040254, 3.141592653589793
])


sin_x_over_x = np.sin(x_values) / x_values

linear_approx = x_values

# Plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, np.sin(x_values), 'b-', label='Analytic (sin(x))')
plt.plot(x_values, linear_approx, 'red', label='Linear Approximation (x)')
plt.plot(x_values, sin_x_over_x, 'g--', label='sin(x)/x Approximation')
plt.grid(True)
plt.title('Taylor Series Approximations of sin(x) at a=0')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.5)  
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.5)  
plt.ylim(-2, 3)
```
