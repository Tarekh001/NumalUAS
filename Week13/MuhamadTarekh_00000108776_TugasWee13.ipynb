```python
import numpy as np
import matplotlib.pyplot as plt

a = 0 
b = 1
n = 4
h = (b - a) / n  
ya = 1.1752
yb = 10.0179

x = np.linspace(a, b, n + 1)

A = np.zeros((n + 1, n + 1))
b_vec = np.zeros(n + 1)

A[0, 0] = 1
b_vec[0] = ya  
A[n, n] = 1
b_vec[n] = yb

for i in range(1, n):
    A[i, i - 1] = 1
    A[i, i] = -(2 + 4 * h**2)  
    A[i, i + 1] = 1
    b_vec[i] = 0


y = np.linalg.solve(A, b_vec)


print("Solusi numerik pada titik-titik grid:")
for i in range(len(x)):
    print(f"x = {x[i]:.2f}, y = {y[i]:.8f}")


plt.plot(x, y, 'o-', label='Solusi Numerik')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solusi BVP: y\'\' = 4y dengan Finite Difference Method')
plt.grid(True)
plt.legend()
plt.show() 
```
